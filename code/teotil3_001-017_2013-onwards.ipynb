# TEOTIL3 data for Oslofjorden

Get TEOTIL3 data for vassdragsområder around Oslofjord.

```python
st_yr, end_yr = 2013, 2023
agri_loss_model = "annual"
vassoms = [f"{id:03d}." for id in range(1, 18)]
```

```python
nve_data_year = end_yr + 1
csv_path = f"/home/jovyan/shared/common/teotil3/evaluation/teo3_results_nve{nve_data_year}_{st_yr}-{end_yr}_agri-{agri_loss_model}-loss.csv"
df = pd.read_csv(csv_path).query("regine in @vassoms")
acc_cols = [col for col in df.columns if col.split("_")[0] == "accum"]
cols = ["regine", "year"] + acc_cols
df = df[cols]
df = df.sort_values(["regine", "year"]).reset_index(drop=True)
df.to_excel(f"../data/oslofjord_summary/teotil3_001-017_{st_yr}-{end_yr}.xlsx", index=False)

df.head()
```
