# TEOTIL2 data for Oslofjorden

Get TEOTIL2 data for vassdragsområder around Oslofjord.

```python
st_yr, end_yr = 1990, 2022
years = range(st_yr, end_yr + 1)
catch_ids = [f"{id:03d}." for id in range(1, 18)]
df_list = []
for year in years:
    df = pd.read_csv(
        f"https://raw.githubusercontent.com/NIVANorge/teotil2/main/data/norway_annual_output_data/teotil2_results_{year}.csv"
    )
    df = df.query("regine in @catch_ids")
    df["year"] = year
    df_list.append(df)

df = pd.concat(df_list, axis="rows")
acc_cols = [col for col in df.columns if col.split("_")[0] == "accum"]
cols = ["regine", "year"] + acc_cols
df = df[cols]
df.sort_values(["regine", "year"], inplace=True)
df.to_excel(f"../data/oslofjord_summary/teotil2_001-017_{st_yr}-{end_yr}.xlsx", index=False)

df.head()
```
