# TEOTIL2-3 data for Oslofjorden

Read combined, bias-corrected TEOTIL2 and TEOTIL3 data for vassdragsområder around Oslofjord.

Based on the workflow used for TEOTIL3 annual reporting.

```python
st_yr, end_yr = 1990, 2023
agri_loss_model = "annual"
vassoms = list(range(1, 18))
```

```python
csv_path = f"/home/jovyan/projects/teotil3_reporting/report_data/{end_yr}/teotil2-3_vassom_npc_{st_yr}-{end_yr}_agri-{agri_loss_model}-loss_bias-corrected.csv"
df = pd.read_csv(csv_path).query("Vassom in @vassoms")
df.to_excel(
    f"../data/oslofjord_summary/teotil2-3_001-017_{st_yr}-{end_yr}.xlsx",
    index=False,
    sheet_name="Tonn",
)

df.head()
```
